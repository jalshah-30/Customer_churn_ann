```python
import tensorflow as tf
from tensorflow.keras.models import load_model
import pickle
import pandas as pd
import numpy as np
```

```python
##Load the trained_model,scalar,le,ohe pickle
model=load_model('model.h5')

#Load encoders
with open('label_encoder_gender.pkl','rb') as file:
    label_encoder_gender=pickle.load(file)

with open('onehot_encoder_geo.pkl','rb') as file:
    onehot_encoder_geo=pickle.load(file)

with open('scaler.pkl','rb') as file:
    scalar=pickle.load(file)
```

```python
# Example input data
input_data = {
    'CreditScore': 600,
    'Geography': 'France',
    'Gender': 'Male',
    'Age': 40,
    'Tenure': 3,
    'Balance': 60000,
    'NumOfProducts': 2,
    'HasCrCard': 1,
    'IsActiveMember': 1,
    'EstimatedSalary': 50000
}
```

```python
geo_encoded = onehot_encoder_geo.transform([[input_data['Geography']]]).toarray()
geo_encoded_df = pd.DataFrame(geo_encoded, columns=onehot_encoder_geo.get_feature_names_out(['Geography']))
```

```python
input_df=pd.DataFrame([input_data])
input_df
```

```python
input_data=pd.concat([input_df.reset_index(drop=True),geo_encoded_df],axis=1)
```

```python
input_data
```

```python
## Encode categorical variables
input_data['Gender']=label_encoder_gender.transform(input_data['Gender'])
input_data
```

```python
input_data.drop('Geography',axis=1,inplace=True)
```

```python
#Scaling the input data
input_scaled=scalar.transform(input_data)
```

```python
prediction=model.predict(input_data)
```

```python
prediction_proba=prediction[0][0]
```

```python
prediction_proba
```

```python
if prediction_proba > 0.5:
    print('The customer is likely to churn.')
else:
    print('The customer is not likely to churn.')
```
